Audio to text using whisper

```python
import whisper

# Model Options
# [tiny, base, small,	medium, large, large-v2]

model = whisper.load_model("base")
result = model.transcribe("/content/Q3.wav")
```

```python
result['text']
```

text to text translation


```python
#I am using googletrans library for translation and installing the ackage compatible with colab
!pip install googletrans==4.0.0-rc1
```

```python
from googletrans import Translator, LANGUAGES

print(LANGUAGES)  # Print supported languages

translator = Translator()
result = translator.translate(result['text'], dest='japanese')  # 'ja' is the language code for Japanese
print(result.text)
```

text to speech

```python
import IPython
import torchaudio
from tortoise.api import TextToSpeech
```

```python
tts = TextToSpeech()
```

```python
# Pick a "preset mode" to determine quality. Options: {"ultra_fast", "fast" (default), "standard", "high_quality"}.
preset = "ultra_fast"
```

```python
gen = tts.tts_with_preset(result.text, voice_samples=None, conditioning_latents=None, preset=preset)
torchaudio.save('generated.wav', gen.squeeze(0).cpu(), 24000)
IPython.display.Audio('generated.wav')
```
